````python
import numpy as np
from matplotlib import pyplot as plt
import math
%matplotlib inline
````

````python
x = np.arange(0, math.pi*4, 0.01)

y1 = np.sin(x)
y2 = np.cos(x)

print('x:', x[:5])
print('sin(x):', y1[:5])
print('cos(x):', y2[:5])
````

````python
plt.plot(x, y1)
plt.plot(x, y2)
````
